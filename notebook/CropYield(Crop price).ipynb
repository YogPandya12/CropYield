```python
## Lets predict CROP_PRICE
features=df.drop(['CROP_PRICE'], axis=1)
target=df['CROP_PRICE']
```

```python
## getting all different of features
num_features= [feature for feature in features.columns if features[feature].dtype!='O']
print('Num of numerical features:',len(num_features))
cat_features=[feature for feature in features.columns if features[feature].dtype=='O']
print('Num of categorical features:',len(cat_features))
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer

# Define the transformers
numeric_transformer = StandardScaler()
ordinal_transformer = OrdinalEncoder()

# Create a ColumnTransformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('OrdinalEncoder', ordinal_transformer, ['STATE', 'SOIL_TYPE', 'CROP']),
        ('StandardScaler', numeric_transformer, ['N_SOIL', 'P_SOIL', 'K_SOIL', 'TEMPERATURE', 'HUMIDITY', 'ph', 'RAINFALL']),
    ],
    remainder='passthrough'  # Keep the rest of the columns unchanged
)

# Assuming 'features' contains your feature data
features_scaled = preprocessor.fit_transform(features)
```

```python
ex=features['CROP']
ex=ordinal_transformer.fit_transform(ex.values.reshape(-1,1))
unique_values = pd.Series(ex.flatten()).unique()
print(unique_values)
```

```python
pd.DataFrame(features_scaled)
```

```python
## separate train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features_scaled,target,test_size=0.2,random_state=42)
x_train.shape,x_test.shape
```

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
import numpy as np

# Convert y_train to a numpy array if it's not already
y_train = np.array(y_train)

# Count the number of samples in each class
class_counts = Counter(y_train)
print(f'Class counts before resampling: {class_counts}')

# Initialize Random Over-Sampler
ros = RandomOverSampler(random_state=42)

# Fit and transform the training data
x_train_resampled, y_train_resampled = ros.fit_resample(x_train, y_train)

# Check the distribution of classes after resampling
print(f'Class counts after resampling: {Counter(y_train_resampled)}')
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV, MultiTaskLassoCV, ElasticNet, ElasticNetCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
```

```python
## create a function to evalute model
def evalute_model(true, predicted):
    mae=mean_absolute_error(true,predicted)
    mse=mean_squared_error(true,predicted)
    rmse=np.sqrt(mean_squared_error(true,predicted))
    r2_square=r2_score(true,predicted)
    return mae, mse, rmse, r2_square
```

```python
## beginning the model training

models = {
    'Linear Regression': (LinearRegression()),
    'Ridge': (Ridge()),
    'ElasticNet': (ElasticNet()),
    'ElasticNetCV': ElasticNetCV(),
    'K-Neighbors Regressor': (KNeighborsRegressor()),
    'Decision Tree': (DecisionTreeRegressor()),
    "Random Forest": (RandomForestRegressor()),
    "Xgboost": (XGBRegressor()),
    'Gradient Boosting': (GradientBoostingRegressor()),
    'AdaBoost': (AdaBoostRegressor())
}


for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train_resampled,y_train_resampled)

    # make predictions
    y_train_pred= model.predict(x_train_resampled)
    y_test_pred= model.predict(x_test)

    model_train_mae,model_train_mse,model_train_rmse,model_train_r2=evalute_model(y_train_resampled,y_train_pred)
    model_test_mae,model_test_mse,model_test_rmse,model_test_r2=evalute_model(y_test,y_test_pred)

    print(list(models.keys())[i])
    print('Model performance for Training set')
    print('Root Mean Squared Error:',model_train_rmse)
    print('Mean Absolute Error:',model_train_mae)
    print('R2 Score:',model_train_r2)
    print()
    print('Model performance for Test set')
    print('Root Mean Squared Error:',model_test_rmse)
    print('Mean Absolute Error:',model_test_mae)
    print('R2 Score:',model_test_r2)
    print('='*30)
```

```python
## Working on different parameters for these models
xgb_params = {
    'learning_rate': [0.01, 0.1],
    'max_depth': [5, 12, 20],
    'n_estimators': [100, 200],
    'colsample_bytree': [0.5, 0.8]
}
rf_params={'max_depth':[5,8,15,None,10],
            'max_features':[5,7,"auto",8],
            'min_samples_split':[2,8,15,20],
            'n_estimators':[100,200,500,1000] }
encv_params = {
    'alpha': [1e-4, 1e-3, 1e-2, 1e-1, 1, 10, 100],
    'l1_ratio': [0, 0.1, 0.5, 0.9, 1],
    'tol': [1e-4, 1e-5, 1e-6],
    'selection': ['cyclic', 'random'],
    'warm_start': [True, False],
    'fit_intercept': [True, False]
}
```

```python
## model list for hyperparameter tuning
models=[
    ('XGB',XGBRegressor(),xgb_params),
    ('RF',RandomForestRegressor(),rf_params),
    ('ENCV',ElasticNetCV(),encv_params)
]
```

```python
from sklearn.model_selection import GridSearchCV

model_param={}
for name, model, params in models:
    # Correctly construct the param_grid for GridSearchCV
    param_grid = {f'{key}': value for key, value in params.items()}

    random=GridSearchCV(estimator=model,
                             param_grid=param_grid,
                             cv=5,
                             verbose=3,
                             n_jobs=-1,
                            scoring='neg_mean_squared_error')
    random.fit(x_train_resampled,y_train_resampled)
    model_param[name]=random.best_params_

for model_name in model_param:
    print(f"Best Params for {model_name}")
    print(model_param[model_name])
```
